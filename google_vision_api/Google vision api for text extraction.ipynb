Changing the PDF file to IMAGE file

```python
#Import the libraries
import io
from PIL import Image
import pytesseract
from pdf2image import convert_from_path
from pytesseract import image_to_string
import os, io
from google.cloud import vision
from google.cloud.vision import types
import pandas as pd

# Conversion of PDF to Image
pages = convert_from_path(r"Cozzini, LLC .Gerardo Arellano 11.09.pdf", 500)
for page in pages:
    page.save('output.png', 'PNG')
    
#Read google vision service account API key 
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = r'naturalanguageapi-8b48acdde8e7.json' 
client = vision.ImageAnnotatorClient()

# Define the file path
FILE_PATH ='./output.png'
with io.open(FILE_PATH, 'rb') as image_file:
    content = image_file.read()
image = vision.types.Image(content=content) 

# Extracting the text from image using document text detection
response = client.document_text_detection(image=image)
docText = response.full_text_annotation.text
print(docText)
"""pages = response.full_text_annotation.pages
for page in pages:
    for block in page.blocks:
        print('block confidence:', block.confidence)
        for paragraph in block.paragraphs:
            print('paragraph confidence:', paragraph.confidence)
            for word in paragraph.words:
                word_text = ''.join([symbol.text for symbol in word.symbols])
                print('Word text: {0} (confidence: {1}'.format(word_text, word.confidence))
                for symbol in word.symbols:
                    print('\tSymbol: {0} (confidence: {1}'.format(symbol.text, symbol.confidence))"""
```
